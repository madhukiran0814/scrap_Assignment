```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd


driver=webdriver.Chrome("chromedriver.exe")

driver.get('https://qcpi.questcdn.com/cdn/posting/?group=1950787&provider=1950787')


table = driver.find_element(By.ID,'table_id')

all_cols=['Post Date','Quest Number','Category Code','Bid/Request Name','Bid Closing Date','City','County','State','Owner','Solicitor','Posting Type']
df = pd.DataFrame()

all_rows = table.find_elements(By.TAG_NAME,'tr')

for row in range(10):
    try:
        col=0
        for tdata in all_rows[row].find_elements(By.TAG_NAME,'td'):
            df.at[row,all_cols[col]]=tdata.text
            col+=1
    except:
        break
df
```
